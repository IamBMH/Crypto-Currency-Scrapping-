Al igual que hemos hecho con la con la API de Coinbase, vamos a proceder a introducir nuestras claves de forma segura con la librería getpass, para el proceso de autentificación. 

```python
# Introducior claves mediante getpass() para no exponer nuestras contraseñas:
consumer_key= getpass("Introduce consumer_key: ")
consumer_secret= getpass("Introduce consumer_secret: ")
access_token= getpass("Introduce access_token: ")
access_token_secret= getpass("Introduce access_token_secret: ")
```

Para proteger los datos de los usuarios, Twitter propone diversas formas de autentificación, en este caso se ha utilizado al autentificación por el método **OAuth**, véase documentación de [Tweepy ](https://https://docs.tweepy.org/en/stable/code_snippet.html#oauth)o documentación de [TwitterAPI](https://developer.twitter.com/en/docs/authentication/overview). El proceso es el siguiente, creamos un objeto que contendrá nuestras claves de acceso, y a esas claves les aplicamos dos [tokens](https://developer.twitter.com/en/docs/authentication/oauth-1-0a/obtaining-user-access-tokens) de autentificación. Una vez creados estos objetos, crearemos el objeto api que nos servirá para realizar la llamada a la API de twitter con la librería tweepy.

```python
# Creamos la conexión con la API con nuestras credenciales
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
# Atenticamos y nos conectamos a la API
api = tweepy.API(auth)
```

La limitaciones que tiene la API de twitter, es que sólo te permite extraer 15000 registros cada 15 minutos, por lo tanto significa que la ejecución de este script tiene que programarse cada 15 minutos, podemos realizar llamadas a a la API de twitter en intervalos de 15 minutos si tenemos pensado automatizar este script. Para tener controladas las llamadas api, hay dos parámetros mostrados a continuación que permiten notificarte y controlar las peticiones realizadas.

```python
# Limitador de llamadas, para no sobrecargar la conexión con la API
api = tweepy.API(auth,wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

Para finalizar con la configuración del filtro, debemos crear un Dataframe vació en el que poder almacenar los datos extraídos, la librería Pandas nos permite crear este tipo de estructuras y modelar los datos sea cual sea nuestra preferencia, por lo tanto, es necesario definir en este paso cuáles son las variables que vamos a recoger.

```python
# Definimos un nombre y le damos forma al dataframe, definiendo las siguientes columnas
article_info = pd.DataFrame(columns=['Date', 'Time', 'Title', 'Articles', 'Link'])
```

A continuación será más necesario la utilización de algunas funciones para ir buscando las noticias en google y para ir almacenando los datos en nuestro dataframe de pandas. Para poder avanzar, se ha definido una función que toma dos parámetros, el primero de ellos toma el dataframe que hemos definido en el paso anterior, y el segundo toma un diccionario con los datos en bruto, la finalidad de esta función es que le vaya llenando dando forma poco a poco al dataframe vació que hemos creado, con cada iteración.

```python
def news_feed(article_info, raw_dictionary):
  for i in range(len(raw_dictionary)-1):
    if raw_dictionary is not None:
      # Busca la fecha y la hora y los convierte a formato DataFrame
      date = raw_dictionary[i]['datetime']
      date = pd.to_datetime(date)
      # Busca el título, hora, descripción y fuente del artículo
      title = raw_dictionary[i]['title']
      time = raw_dictionary[i]['date']
      articles = raw_dictionary[i]['desc']
      link = raw_dictionary[i]['link']
      # Unimos toda la información anterior de las variables anteriores
      article_info = article_info.append({'Date':date, 'Time': time, 'Title': title,
                                          'Articles': articles, 'Link': link}, ignore_index=True)
    else:
        break
  return article_info
```

```python
# Dataframe vació que va a ir almacenando los datos en bruto
articles = pd.DataFrame()

# Cada Keyword será buscada por separado
for steps in range(len(keywords)):
  string = (keywords[steps])
  googlenews.search(string)

  #Buscar los resultados en Google news sobre un tema en específico y lo almacena en un variable
  result = googlenews.results()


  #Queremos los resultados de la primera página, los más relevantes y los que más impacto tiene a nivel global en el mercado moderno, por lo tanto fijamos el parámetro total_pages en 1
  total_pages = 1 # Podemos aumentar el número de páginas

  for steps in range(total_pages):
    # La variable consta de páginas especificadas por el usuario, e usa un bucle 'for' para almacenador todos los datos en el DataFrame
    googlenews.get_page(steps)
    feed = news_feed(article_info, result)

  articles = articles.append(feed)

  # Es necesario limpiar los resultados de búsqueda de las anteriores keywords para evitar datos duplicados
  googlenews.clear()

shape = articles.shape[0]

# Guardamos los datos de la búqueda en equipo local en caso de que estemos trabajando en jupyter notebook
#articles.to_csv('C:/Users/bruno/Desktop/CryptoDatasets/GoogleNews.csv', index=False)



# Guardamos el dataset en nuestro equipo local en case de que estemos trabajando con Google Colab
articles.to_csv('GoogleNews.csv')
from google.colab import files
files.download("GoogleNews.csv")



#Función para resetear el índice del resultado final
articles.index = np.arange(shape)
articles
```
